```python
x = dt[['age', 'sex', 'bmi', 'children', 'smoker', 'region']]
y = dt['charges']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.2, random_state = 2)
```

```python
lz = LazyRegressor(predictions = True)
models, predictions = lz.fit(xtrain, xtest, ytrain, ytest)
```
